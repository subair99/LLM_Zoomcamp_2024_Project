```python
# Import the required modules.
import pandas as pd
```

```python
# Read the data.
df = pd.read_csv('./medquad.csv')
```

```python
# Show the head of the df.
df.head()
```

```python
# Show the shape of the data.
df.shape
```

```python
# Show the counts of the data.
df.source.value_counts()
```

```python
# Remove unwanted spaces in the columns.
df.question = df.question.str.replace(' ?', '?')
```

```python
# Remove unwanted spaces in the columns.
df.question = df.question.str.replace('??', '?')
```

```python
# Remove row with duplicate questions.
df = df.drop_duplicates(subset='question')
```

```python
# Show the shape df.
df.shape
```

```python
# Find the list of the sources in df.
source_list = list(set(df.source))
```

```python
# Create an empty list to store the index.
empty_list = []
```

```python
# Create the dataframe of the selected indices.
for item in source_list:
    index_list = df[df.source==item].sample(n=20, random_state=2).index.to_list()

    for num in index_list:
        empty_list.append(num)

empty_list.sort()
df_index = df.loc[empty_list]
```

```python
# Show the head of df_index.
df_index.head()
```

```python
# Check the length of df_index.
df_index.shape
```

```python
# Re-index df_index.
df_index = df_index.reset_index(drop=True)
```

```python
# Create an id column with the index.
id_column = list(range(len(df_index)))
df_index.loc[:, 'id'] = id_column
```

```python
# Show the head of df_index.
df_index.head()
```

```python
# Rename focus_area to focus.
df_index.columns = [['question', 'answer', 'source', 'focus', 'id']]
```

```python
# Show the head of df_index.
df_index.head()
```

```python
# Re-arrange df.
df_index = df_index[['id', 'question', 'answer', 'focus', 'source']]
```

```python
# Show the head of df_index.
df_index.head()
```

```python
# Save df_index to disk as data.csv.
df_index.to_csv('./data.csv', index=False)
```
